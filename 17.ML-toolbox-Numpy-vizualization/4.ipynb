### Q1. Load the "titanic" dataset using the `load_dataset` function of seaborn. Use Plotly express to plot a scatter plot for `age` and `fare` columns in the titanic dataset.

```python
import seaborn as sns
import plotly.express as px

# Load the titanic dataset
titanic = sns.load_dataset('titanic')

# Plot scatter plot for 'age' vs 'fare'
fig = px.scatter(titanic, x='age', y='fare', title='Scatter plot of Age vs Fare')
fig.show()
```

### Q2. Using the `tips` dataset in the Plotly library, plot a box plot using Plotly express.

```python
import plotly.express as px

# Load the tips dataset
tips = px.data.tips()

# Plot box plot
fig = px.box(tips, x='day', y='total_bill', title='Box plot of Total Bill across Days')
fig.show()
```

### Q3. Using the `tips` dataset in the Plotly library, plot a histogram for `x= "sex"` and `y="total_bill"` column in the tips dataset. Also, use the "smoker" column with the `pattern_shape` parameter and the "day" column with the `color` parameter.

```python
import plotly.express as px

# Load the tips dataset
tips = px.data.tips()

# Plot histogram
fig = px.histogram(tips, x='sex', y='total_bill', color='day', pattern_shape='smoker', title='Histogram of Total Bill by Sex with Day and Smoker')
fig.show()
```

### Q4. Using the `iris` dataset in the Plotly library, plot a scatter matrix plot, using the "species" column for the `color` parameter.

```python
import plotly.express as px

# Load the iris dataset
iris = px.data.iris()

# Plot scatter matrix
fig = px.scatter_matrix(iris, dimensions=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'], color='species', title='Scatter Matrix of Iris Dataset')
fig.show()
```

### Q5. What is a `Distplot`? Using Plotly express, plot a distplot.

A `distplot` (distribution plot) shows the distribution of a univariate set of data, often combined with a histogram and kernel density estimate (KDE). It provides insights into the data's spread, skewness, and modality.

```python
import plotly.express as px

# Load the tips dataset
tips = px.data.tips()

# Plot distplot (using histogram with KDE)
fig = px.histogram(tips, x='total_bill', nbins=30, marginal='violin', title='Distplot of Total Bill')
fig.show()
```
