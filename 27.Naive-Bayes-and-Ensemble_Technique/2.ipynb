## Q1. A company conducted a survey of its employees and found that 70% of the employees use the company's health insurance plan, while 40% of the employees who use the plan are smokers. What is the probability that an employee is a smoker given that he/she uses the health insurance plan?

We are asked to calculate \( P(\text{Smoker} | \text{Health Insurance}) \).

Given:
- \( P(\text{Health Insurance}) = 0.70 \)
- \( P(\text{Smoker} | \text{Health Insurance}) = 0.40 \)

Thus, the probability that an employee is a smoker given that he/she uses the health insurance plan is:
\[
P(\text{Smoker} | \text{Health Insurance}) = 0.40
\]

## Q2. What is the difference between Bernoulli Naive Bayes and Multinomial Naive Bayes?

- **Bernoulli Naive Bayes**: This is used when the features are binary (0 or 1). It considers each feature as independent and represents whether a particular attribute is present or absent in a sample.
  
- **Multinomial Naive Bayes**: This is used for multinomially distributed data, typically for discrete features such as word counts in a document. It is most suitable for text classification problems where the frequency of each feature (such as word occurrence) is important.

The key difference is that Bernoulli Naive Bayes is for binary/boolean data, while Multinomial Naive Bayes is for count-based or categorical data.

## Q3. How does Bernoulli Naive Bayes handle missing values?

Bernoulli Naive Bayes does not handle missing values explicitly. In cases where there are missing values in the dataset, preprocessing is required, such as imputing missing values or removing data points with missing values. The classifier assumes all features are binary, so missing values need to be appropriately handled during data preparation.

## Q4. Can Gaussian Naive Bayes be used for multi-class classification?

Yes, **Gaussian Naive Bayes** can be used for multi-class classification. It extends naturally to multi-class problems by applying Bayes' theorem independently to each class and selecting the class with the highest posterior probability. Gaussian Naive Bayes works well with continuous features and assumes that the features follow a normal (Gaussian) distribution.
