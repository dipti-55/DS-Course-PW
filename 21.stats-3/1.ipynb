| **Criterion**          | **t-test**                                              | **z-test**                                               |

### Example Scenarios

**Scenario:** You are a researcher studying the average test scores of a new teaching method. You have a sample of 20 students and their average score is 78. You want to compare this to the national average test score of 75, but the population variance is unknown.

**t-test Formula:**
t = (x̄ - μ) / (s / √n)
where:
- x̄ = 78 (sample mean)
- μ = 75 (population mean)
- s is the sample standard deviation
- n = 20 (sample size)

#### z-test Example
**Scenario:** A pharmaceutical company claims that their drug improves recovery time from a particular illness. The population mean recovery time is 8 days with a known standard deviation of 1.5 days. You have a sample of 50 patients who took the drug, and their average recovery time is 7.5 days.

z = (x̄ - μ) / (σ / √n)
where:
- x̄ = 7.5 (sample mean)

- σ = 1.5 (population standard deviation)
- n = 50 (sample size)

### Summary

## Q2: Differentiate between one-tailed and two-tailed tests.

### One-tailed vs. Two-tailed Tests

| **Criterion**          | **One-tailed Test**                                   | **Two-tailed Test**                                    |
|------------------------|-------------------------------------------------------|-------------------------------------------------------|
| **Purpose**            | Tests for a deviation in one specific direction       | Tests for a deviation in either direction              |
| **Hypotheses**         | - Null hypothesis (\(H_0\)): No effect or relationship<br>- Alternative hypothesis (\(H_a\)): Effect or relationship in one direction (e.g., greater than or less than) | - Null hypothesis (\(H_0\)): No effect or relationship<br>- Alternative hypothesis (\(H_a\)): Effect or relationship in either direction (e.g., not equal to) |
| **Critical Region**    | One side of the distribution                          | Both sides of the distribution                         |
| **Significance Level** | Entire alpha level (e.g., 0.05) in one tail           | Alpha level split between two tails (e.g., 0.025 in each tail if alpha is 0.05) |
| **Usage**              | When the research hypothesis predicts a direction     | When the research hypothesis does not predict a direction or is exploratory |

### Example Scenarios

#### One-tailed Test Example
**Scenario:** A company claims that a new training program increases employee productivity. You want to test if the average productivity after the training is significantly greater than the previous average productivity of 70 units.

- **Null hypothesis (\(H_0\))**: The average productivity after training is less than or equal to 70 units.
- **Alternative hypothesis (\(H_a\))**: The average productivity after training is greater than 70 units.

Here, you are only interested in testing if the productivity has increased, so you use a one-tailed test.

#### Two-tailed Test Example
**Scenario:** A pharmaceutical company wants to test if a new drug has a different effect on blood pressure compared to the existing drug, which has an average effect of reducing blood pressure by 10 mmHg.

- **Null hypothesis (\(H_0\))**: The new drug's effect on blood pressure is equal to 10 mmHg.
- **Alternative hypothesis (\(H_a\))**: The new drug's effect on blood pressure is not equal to 10 mmHg.

Here, you are interested in testing if the new drug's effect is either greater than or less than 10 mmHg, so you use a two-tailed test.

### Summary
- **One-tailed Test**: Used when the hypothesis predicts the direction of the effect (e.g., greater than or less than).
- **Two-tailed Test**: Used when the hypothesis does not predict the direction of the effect or is exploratory (e.g., not equal to).

## Q3: Explain the concept of Type 1 and Type 2 errors in hypothesis testing. Provide an example scenario for each type of error.

### Type 1 and Type 2 Errors in Hypothesis Testing

In hypothesis testing, we can make two types of errors:

#### Type 1 Error (False Positive)
- **Definition**: Occurs when the null hypothesis (\(H_0\)) is true, but we incorrectly reject it.
- **Probability**: Denoted by alpha (α), which is the significance level of the test. Common values are 0.05, 0.01, etc.
- **Consequence**: Concluding that there is an effect or difference when there actually isn't one.

**Example Scenario:**
Suppose a pharmaceutical company tests a new drug to see if it is more effective than the existing drug. The null hypothesis (\(H_0\)) is that the new drug has no difference in effectiveness compared to the existing drug.

- **Null hypothesis (\(H_0\))**: The new drug is not more effective than the existing drug.
- **Alternative hypothesis (\(H_a\))**: The new drug is more effective than the existing drug.

A Type 1 error would occur if the test results lead the researchers to conclude that the new drug is more effective (rejecting \(H_0\)), when in reality, it is not more effective.

#### Type 2 Error (False Negative)
- **Definition**: Occurs when the null hypothesis (\(H_0\)) is false, but we fail to reject it.
- **Probability**: Denoted by beta (β). The power of the test is 1 - β, which represents the probability of correctly rejecting a false null hypothesis.
- **Consequence**: Concluding that there is no effect or difference when there actually is one.

**Example Scenario:**
Using the same pharmaceutical company example, suppose the null hypothesis (\(H_0\)) is that the new drug has no difference in effectiveness compared to the existing drug.

- **Null hypothesis (\(H_0\))**: The new drug is not more effective than the existing drug.
- **Alternative hypothesis (\(H_a\))**: The new drug is more effective than the existing drug.

A Type 2 error would occur if the test results lead the researchers to conclude that there is no difference in effectiveness (failing to reject \(H_0\)), when in reality, the new drug is more effective.

### Summary of Errors

| **Type of Error** | **Definition**                                                           | **Consequences**                                                                                     | **Example**                                                                                          |
|-------------------|--------------------------------------------------------------------------|------------------------------------------------------------------------------------------------------|------------------------------------------------------------------------------------------------------|
| **Type 1 Error**  | Rejecting the null hypothesis when it is true (False Positive)           | Concluding there is an effect or difference when there is none                                        | Concluding a new drug is effective when it is not                                                    |
| **Type 2 Error**  | Failing to reject the null hypothesis when it is false (False Negative)  | Concluding there is no effect or difference when there is one                                         | Concluding a new drug is not effective when it actually is                                            |

### Visual Representation
Imagine the null hypothesis as a courtroom scenario:

- **Type 1 Error**: Convicting an innocent person (rejecting \(H_0\) when true).
- **Type 2 Error**: Letting a guilty person go free (failing to reject \(H_0\) when false).

Balancing these errors is crucial in hypothesis testing, as reducing the probability of one type of error typically increases the probability of the other.


## Q4:  Explain Bayes's theorem with an example.

### Bayes's Theorem

Bayes's Theorem is a way of finding a probability when we know certain other probabilities. It is stated as:

 P(A|B) = (P(B|A).P(A))/P(B)

where:
- \( P(A|B) \) is the probability of event A occurring given that B is true.
- \( P(B|A) \) is the probability of event B occurring given that A is true.
- \( P(A) \) is the probability of event A.
- \( P(B) \) is the probability of event B.

### Example Scenario

**Scenario**: Medical Testing

Let's say there is a disease that affects 1% of the population. There is a test for the disease that is 99% accurate:
- If a person has the disease, the test will correctly identify it (true positive) 99% of the time.
- If a person does not have the disease, the test will correctly identify that they do not have it (true negative) 99% of the time.

Now, suppose a person tests positive for the disease. We want to find the probability that they actually have the disease.

**Step-by-step Solution Using Bayes's Theorem:**

1. **Define the Events:**
   - \( A \): The event that the person has the disease.
   - \( B \): The event that the person tests positive.

2. **Given Probabilities:**
   - \( P(A) = 0.01 \) (1% of the population has the disease)
   - \( P(not A) = 0.99 \) (99% of the population does not have the disease)
   - \( P(B|A) = 0.99 \) (Probability of testing positive given the person has the disease)
   - \( P(B|not A) = 0.01 \) (Probability of testing positive given the person does not have the disease, which is the false positive rate)

3. **Calculate \( P(B) \):**
   Using the law of total probability:
   P(B) = P(B∣A)⋅P(A) + P(B∣¬A)⋅P(¬A)
   
   Substituting the values:
   P(B)=(0.99⋅0.01)+(0.01⋅0.99)=0.0099+0.0099=0.0198

4. **Apply Bayes's Theorem:**
    P(A|B) = (P(B|A).P(A))/P(B)

   Substituting the values:
   P(A|B) = (0.99*0.01)/0.0198 = 0.50 approx

So, even if a person tests positive, there is only about a 50% chance that they actually have the disease, given the low prevalence of the disease in the population and the accuracy of the test.

### Summary
Bayes's Theorem helps us update the probability of an event based on new evidence. In this example, it shows that despite a positive test result, the actual probability of having the disease is only 50%, highlighting the importance of considering both the accuracy of the test and the prevalence of the disease in the population.

## Q5: What is a confidence interval? How to calculate the confidence interval, explain with an example.

### Confidence Interval

A confidence interval is a range of values, derived from a sample, that is likely to contain the value of an unknown population parameter. The interval has an associated confidence level that quantifies the level of confidence that the parameter lies within the interval. Common confidence levels include 90%, 95%, and 99%.

### How to Calculate a Confidence Interval

To calculate a confidence interval for a population mean, we typically use the following formula:

Confidence Interval = x̄ ± z* * (s / √n)

where:
- x̄ is the sample mean.
- z* is the z-value that corresponds to the desired confidence level.
- s is the sample standard deviation.
- n is the sample size.

### Example Calculation

Let's go through an example step by step:

**Scenario:** Suppose we want to estimate the average height of adult males in a city. We take a random sample of 100 adult males and find the following:
- Sample mean height (x̄) = 175 cm
- Sample standard deviation (s) = 10 cm
- Sample size (n) = 100

We want to calculate a 95% confidence interval for the population mean height.

**Step-by-step Calculation:**

1. **Determine the z-value for the desired confidence level:**
   - For a 95% confidence level, the z-value (z*) is approximately 1.96 (this value comes from the standard normal distribution table).

2. **Calculate the standard error of the mean:**
   Standard Error = s / √n
   Substituting the values:
   Standard Error = 10 / √100 = 10 / 10 = 1

3. **Calculate the margin of error:**
   Margin of Error = z* * Standard Error
   Substituting the values:
   Margin of Error = 1.96 * 1 = 1.96

4. **Calculate the confidence interval:**
   Confidence Interval = x̄ ± Margin of Error
   Substituting the values:
   Confidence Interval = 175 ± 1.96
   This gives us:
   Confidence Interval = (175 - 1.96, 175 + 1.96) = (173.04, 176.96)

### Interpretation

We can say with 95% confidence that the true mean height of adult males in the city is between 173.04 cm and 176.96 cm.

### Summary

A confidence interval provides a range of plausible values for a population parameter. The wider the interval, the more uncertainty there is about the parameter's value. The confidence level indicates the degree of certainty we have that the interval contains the parameter.

The margin of error in a confidence interval represents the range within which we expect the true population parameter to fall, given the sample data. **Margin of Error Formula:**

    Margin of Error = z * (s / √n)

where:
- `z` is the z-value corresponding to the desired confidence level.
- `s` is the sample standard deviation.
- `n` is the sample size.

### How Sample Size Affects the Margin of Error

The margin of error decreases as the sample size increases. This is because the standard error of the mean (s / √n) decreases as n increases. In other words, a larger sample size provides a more accurate estimate of the population parameter, which results in a smaller margin of error.

## Q12. A researcher is testing the effectiveness of two different teaching methods on student performance. To determine if there is a significant difference in student performance between the two teaching methods, we can conduct an independent two-sample t-test. Here’s a step-by-step guide:

### Hypothesis Test Setup

**1. Define Hypotheses:**

- **Null Hypothesis (H0):** There is no difference in mean scores between the two teaching methods.
  \[
  H0: μA = μB
  \]

- **Alternative Hypothesis (H1):** There is a difference in mean scores between the two teaching methods.
  \[
  H1: μA ≠ μB
  \]

**2. Given Data:**

- Sample A:
  - Mean (x̄A) = 85
  - Standard deviation (sA) = 6
  - Sample size (nA) = Assume nA = 30 for this example

- Sample B:
  - Mean (x̄B) = 82
  - Standard deviation (sB) = 5
  - Sample size (nB) = Assume nB = 30 for this example

- Significance level (α) = 0.01

**3. Calculate the Test Statistic:**

The test statistic for an independent two-sample t-test is calculated using the formula:

    t = (x̄A - x̄B) / √((sA^2 / nA) + (sB^2 / nB))

where:
- x̄A = 85
- x̄B = 82
- sA = 6
- sB = 5
- nA = 30
- nB = 30

Substituting the values:

    t = (85 - 82) / √((6^2 / 30) + (5^2 / 30))
      = 3 / √((36 / 30) + (25 / 30))
      = 3 / √(1.2 + 0.833)
      = 3 / √2.033
      = 3 / 1.426
      ≈ 2.10

**4. Determine the Degrees of Freedom:**

Degrees of freedom for the t-test can be approximated using:

    df ≈ ((sA^2 / nA) + (sB^2 / nB))^2 / [((sA^2 / nA)^2 / (nA - 1)) + ((sB^2 / nB)^2 / (nB - 1))]

Substituting the values:

    df ≈ ((1.2 + 0.833)^2) / [((1.2^2 / 29) + (0.833^2 / 29))]
       ≈ (2.033^2) / [(0.050 + 0.024)]
       ≈ 4.13 / 0.074
       ≈ 55.8

We approximate df to 56.

**5. Find the Critical t-value:**

For a two-tailed test at α = 0.01 and df ≈ 56, the critical t-value from the t-distribution table is approximately ±2.673.

**6. Compare the Test Statistic to the Critical t-value:**

- The calculated t-value is approximately 2.10.
- The critical t-value is ±2.673.

**7. Conclusion:**

Since the calculated t-value (2.10) is less than the critical t-value (±2.673), we fail to reject the null hypothesis. This means there is not enough evidence to conclude that there is a significant difference in student performance between the two teaching methods at the 0.01 significance level.

### Summary

Based on the t-test, we do not have sufficient evidence to say that the two teaching methods result in significantly different student performance at the 0.01 significance level.

## Q14. In a study of the effects of caffeine on reaction time, a sample of 30 participants had an average reaction time of 0.25 seconds with a standard deviation of 0.05 seconds. Conduct a hypothesis test to determine if the caffeine has a significant effect on reaction time at a 90% confidence level using a t-test.

To determine if caffeine has a significant effect on reaction time, we can perform a one-sample t-test. Here’s a step-by-step guide for conducting the hypothesis test:

### Hypothesis Test Setup

**1. Define Hypotheses:**

- **Null Hypothesis (H0):** Caffeine has no effect on reaction time. The average reaction time is equal to the expected average without caffeine.
  \[
  H0: μ = μ0
  \]

- **Alternative Hypothesis (H1):** Caffeine has an effect on reaction time. The average reaction time is not equal to the expected average without caffeine.
  \[
  H1: μ ≠ μ0
  \]

**2. Given Data:**

- Sample mean (x̄) = 0.25 seconds
- Sample standard deviation (s) = 0.05 seconds
- Sample size (n) = 30
- Confidence level = 90%

**3. Expected Average Without Caffeine (μ0):**

To perform the test, we need a comparison value for the average reaction time without caffeine. Assume this value is 0.30 seconds (this value is hypothetical for illustration purposes).

**4. Calculate the Test Statistic:**

The test statistic for a one-sample t-test is calculated using the formula:

    t = (x̄ - μ0) / (s / √n)

where:
- x̄ = 0.25
- μ0 = 0.30 (expected average without caffeine)
- s = 0.05
- n = 30

Substituting the values:

    t = (0.25 - 0.30) / (0.05 / √30)
      = -0.05 / (0.05 / 5.477)
      = -0.05 / 0.0091
      ≈ -5.49

**5. Determine the Degrees of Freedom:**

Degrees of freedom for the t-test is calculated as:

    df = n - 1
       = 30 - 1
       = 29

**6. Find the Critical t-value:**

For a two-tailed test at a 90% confidence level and df = 29, the critical t-value is approximately ±1.699.

**7. Compare the Test Statistic to the Critical t-value:**

- The calculated t-value is approximately -5.49.
- The critical t-value is ±1.699.

**8. Conclusion:**

Since the absolute value of the calculated t-value (5.49) is greater than the critical t-value (1.699), we reject the null hypothesis. This indicates that there is a significant effect of caffeine on reaction time at the 90% confidence level.

### Summary

Based on the t-test, we conclude that caffeine significantly affects reaction time, with the observed average reaction time being significantly different from the expected average without caffeine at the 90% confidence level.
