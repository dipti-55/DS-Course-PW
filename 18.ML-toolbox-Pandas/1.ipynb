### Q1. Create a Pandas Series that contains the following data: 4, 8, 15, 16, 23, and 42. Then, print the series.

### Q3. Create a Pandas DataFrame that contains the following data, then print the DataFrame.
| Name   | Age | Gender |
|--------|-----|--------|
| Alice  | 25  | Female |
| Bob    | 30  | Male   |
| Claire | 27  | Female |
```python
data = {
    'Name': ['Alice', 'Bob', 'Claire'],
    'Age': [25, 30, 27],
    'Gender': ['Female', 'Male', 'Female']
}
df = pd.DataFrame(data)
print(df)
```

### Q4. What is ‘DataFrame’ in pandas, and how is it different from `pandas.Series`? Explain with an example.
- A **DataFrame** in pandas is a two-dimensional, size-mutable, and heterogeneous data structure with labeled axes (rows and columns). It can be thought of as a table or spreadsheet.
- A **Series** in pandas is a one-dimensional array-like object containing an array of data and an associated array of data labels, called its index.

#### Example:
```python
# Creating a Series
series_example = pd.Series([1, 2, 3, 4])
print(series_example)

# Creating a DataFrame
df_example = pd.DataFrame({
    'Numbers': [1, 2, 3, 4],
    'Letters': ['A', 'B', 'C', 'D']
})
print(df_example)
```

### Q5. What are some common functions you can use to manipulate data in a Pandas DataFrame? Can you give an example of when you might use one of these functions?

Some common functions to manipulate data in a DataFrame include:
- `df.head()` – To view the first few rows of the DataFrame.
- `df.tail()` – To view the last few rows.
- `df.describe()` – To get summary statistics.
- `df.drop()` – To drop specified rows or columns.
- `df.fillna()` – To fill missing values.

#### Example:
```python
# Dropping a column
df_modified = df.drop('Gender', axis=1)
print(df_modified)
```

### Q6. Which of the following is mutable in nature: Series, DataFrame, Panel?
- **Series**: Mutable.
- **DataFrame**: Mutable.
- **Panel**: Mutable (although Panels are deprecated as of Pandas version 1.0.0).

### Q7. Create a DataFrame using multiple Series. Explain with an example.
```python
# Creating individual Series
names = pd.Series(['Alice', 'Bob', 'Claire'])
ages = pd.Series([25, 30, 27])
genders = pd.Series(['Female', 'Male', 'Female'])

# Creating a DataFrame from the Series
df_from_series = pd.DataFrame({
    'Name': names,
    'Age': ages,
    'Gender': genders
})

print(df_from_series)
```
